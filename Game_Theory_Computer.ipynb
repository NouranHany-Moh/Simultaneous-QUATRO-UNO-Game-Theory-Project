```python
from itertools import permutations
first_row = permutations("1234")
first_row = list(first_row)
first_row.insert(0, "")
first_row_str = []
first_row_string_=[]
print(first_row)
for i in first_row:
    first_row_string_.append(''.join(i)) 
first_row_str.append(first_row_string_)
print(first_row_string_)#desired to be list of list

for i in first_row_string_:
    if i != '':
        player1_element = i
        next_row = [i]#inner list

        for j in first_row_string_:
            if j !='':
                iter_player1 = 0#penalty
                iter_player2 = 0#penalty
                while iter_player1 < 4 and iter_player2 < 4:
                    if (i[iter_player1] == '1' and j[iter_player2] == '4') or (i[iter_player1] == '4' and j[iter_player2] == '1'):
                        iter_player1 += 1
                        iter_player2 += 1
                    elif i[iter_player1] == j[iter_player2]:
                        iter_player1 += 1
                        iter_player2 += 1
                    elif i[iter_player1] > j[iter_player2]:
                        iter_player2 += 1
                    else:
                        iter_player1 += 1

                if iter_player1 == 4 and iter_player2 == 4: 
                    next_row.append("0")
                elif iter_player1 == 4:
                    next_row.append("-1")
                else:
                    next_row.append("1")

        first_row_str.append(next_row)
print(first_row_str)
```

```python
def write_excel(ann_order,beth_order):
    
    workbook = xlsxwriter.Workbook("C:\\Users\\OP\\Desktop\\Book1.xlsx")
    worksheet = workbook.add_worksheet('Game Theory')
    book_format = workbook.add_format(properties={'bold': True, 'bg_color': 'yellow'})
    for i in range(25):#row
        for j in range(25):#column
            worksheet.write(i, j, first_row_str[i][j])
            if first_row_str[i][0]==ann_order and first_row_str[0][j]==beth_order:
                worksheet.write(i, j, first_row_str[i][j],book_format)

    workbook.close()
```

```python
def results():
    global score_Ann
    global score_Beth
    global draw
    global iteration
    global print_msg
    while not index_ann.empty() and not index_beth.empty():
        if (index_ann.queue[0]==1 and index_beth.queue[0]==4) or(index_beth.queue[0]==1 and index_ann.queue[0]==4):
            #both are destroyed
            index_ann.get()
            index_beth.get()        
        elif index_ann.queue[0]>index_beth.queue[0]:
            #beth is destroyed
            index_beth.get()
            
        elif index_beth.queue[0]>index_ann.queue[0]:
            #ann is destroyed
            index_ann.get()
        
        
        elif index_beth.queue[0]==index_ann.queue[0]:
            #both are destroyed
            index_ann.get()
            index_beth.get()
            
    if index_ann.empty() and index_beth.empty():
        draw+=1
        textbox2.config(text=str(draw))
        label8 = tk.Label(childframe7, text=print_msg+":Draw", bg="#889988", width=0, height=0,font=("Arial", 20, "bold"), anchor='nw',fg="white")
        label8.grid(row=draw+score_Beth+score_Ann, column=0, padx=0, pady=0, sticky="w")
        
    elif index_ann.empty():
        score_Beth+=1
        textbox1.config(text=str(score_Beth))
        label8 = tk.Label(childframe7, text=print_msg+":Beth won", bg="#889988", width=0, height=0,font=("Arial", 20, "bold"), anchor='nw',fg="white")
        label8.grid(row=draw+score_Beth+score_Ann, column=0, padx=0, pady=0, sticky="w")
        
    else:
        score_Ann+=1
        textbox.config(text=str(score_Ann))
        label8 = tk.Label(childframe7, text=print_msg+":Ann Won", bg="#889988", width=0, height=0,font=("Arial", 20, "bold"), anchor='nw',fg="white")
        label8.grid(row=draw+score_Beth+score_Ann, column=0, padx=0, pady=0, sticky="w")
    destroy_remains()
```

```python
def text_printing(my_queue1,my_queue2):
    print_message="playing"
    global final_ann
    global final_beth
    for i in range(4):
        element = str(my_queue1.queue[i])
        final_ann+=str(my_queue1.queue[i])
        print_message += element + " "
    print_message+="aganist"+" "
    for i in range(4):
        element = str(my_queue2.queue[i])
        final_beth+=str(my_queue2.queue[i])
        print_message += element + " "
    
    
    return str(print_message)
```

```python
def show_frame(frame_number,player):
    global iteration
    global order_ann#to intend
    global order_beth#to intend
    global ann_copy
    global beth_copy
    global print_msg
    
    if player=="Ann":
    
        grandchild_frame1 = tk.Frame(child_frame4, highlightbackground="black", highlightthickness=1, height=80, width=50, bg="white")
        grandchild_frame1.grid_propagate(False)  # disable resizing
        grandchild_frame1.grid(row=1, column=order_ann, padx=0, pady=0, columnspan=1, sticky="e")
        label_grad1 = tk.Label(grandchild_frame1, text=frame_number, bg="white", width=0, height=0, fg="black", anchor='nw', justify="left", font=("Arial", 8, "bold"))
        label_grad1.grid(row=0,column=0)
        
        iteration +=1
        order_ann+=1
    
    if player=="Beth":
        grandchild_frame1 = tk.Frame(child_frame6, highlightbackground="black", highlightthickness=1, height=80, width=50, bg="white")
        grandchild_frame1.grid_propagate(False)  # disable resizing
        grandchild_frame1.grid(row=1, column=order_beth, padx=0, pady=0, columnspan=1, sticky="e")
        label_grad1 = tk.Label(grandchild_frame1, text=frame_number, bg="white", width=0, height=0, fg="black", anchor='nw', justify="left", font=("Arial", 8, "bold"))
        label_grad1.grid(row=0,column=0)
        
        iteration+=1
        order_beth+=1  
    if iteration==8:
        ann_copy=copy(index_ann)
        beth_copy=copy(index_beth)
        ann_printing=copy(index_ann)
        beth_printing=copy(index_beth)
        print_msg=text_printing(ann_printing,beth_printing)
           
        clear_frame(child_frame4,"Ann",ann_copy,beth_copy)#if done
```

```python
def change_color(frame):
    frame.config(bg="#889988")
```

```python
def clear_frame(frame,player,ann_copy,beth_copy):
    frame_children4 = [child for child in child_frame4.winfo_children() if isinstance(child, tk.Frame)]
    frame_children6 = [child for child in child_frame6.winfo_children() if isinstance(child, tk.Frame)]
    delay=0
    ann_iter=0
    beth_iter=0
    for i in range(4):
      
        if not ann_copy.empty() and not beth_copy.empty():
            frame_children6[beth_iter].after(delay, change_color,frame_children6[beth_iter])
            frame_children4[ann_iter].after(delay, change_color,frame_children4[ann_iter])
            delay+=2000

            if (ann_copy.queue[0]==beth_copy.queue[0] )or (ann_copy.queue[0]==1 and beth_copy.queue[0]==4)or (beth_copy.queue[0]==1 and ann_copy.queue[0]==4):
                
                frame_children4[ann_iter].after(delay, frame_children4[ann_iter].destroy)
                frame_children6[beth_iter].after(delay, frame_children6[beth_iter].destroy)
                ann_iter+=1
                beth_iter+=1
                ann_copy.get()
                beth_copy.get()

        #             frame_children6[i].destroy()


            elif ann_copy.queue[0]>beth_copy.queue[0]:
              
                frame_children6[beth_iter].after(delay, frame_children6[beth_iter].destroy)
                beth_copy.get()
                beth_iter+=1
            elif ann_copy.queue[0]<beth_copy.queue[0]:
                
                frame_children4[ann_iter].after(delay, frame_children4[ann_iter].destroy)
                ann_copy.get()
                ann_iter+=1
            delay+=2000  
  
    window.after(delay, results) 
           
```

```python
def hide_frame(event,frame_number,player):
    frame_to_hide = event.widget
    frame_to_hide.destroy()
    if player=="Ann":
        index_ann.put(frame_number)
        
    else:
        index_beth.put(frame_number)
    show_frame(frame_number,player)
```

```python
def hide_random_frame():
    frame_children3 = [child for child in child_frame3.winfo_children() if isinstance(child, tk.Frame)]
    delay=0
    ann_iter=0
    beth_iter=0
    my_list = [0,1,2,3]
    delay=4000
    random.shuffle(my_list)
    global iteration
    for i in my_list:
        index_beth.put(i+1)
        frame_children3[i].after(delay, lambda idx=i+1: show_frame(idx, "Beth"))
        frame_children3[i].after(delay,frame_children3[i].destroy)
        delay+=2000
        
        
    
```
